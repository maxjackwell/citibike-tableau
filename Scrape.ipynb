```python
import pandas as pd
```

```python
# Read data from 2023 July files
file1 = '23_July_1.csv'
file2 = '23_July_2.csv'
file3 = '23_July_3.csv'
file4 = '23_July_4.csv'


df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)
df4 = pd.read_csv(file4)
```

```python
df1.head()
```

```python
# List of DataFrame names and the corresponding DataFrame objects
dfs = {'df1': df1, 'df2': df2, 'df3': df3, 'df4': df4}

for df_name, df in dfs.items():
    # Convert 'started_at' and 'ended_at' to datetime if they are not already
    df['started_at'] = pd.to_datetime(df['started_at'])
    df['ended_at'] = pd.to_datetime(df['ended_at'])
    
    # Calculate the time difference in seconds
    df['trip_duration_seconds'] = (df['ended_at'] - df['started_at']).dt.total_seconds()
    
    # Optionally, create a new column for trip duration in minutes
    df['trip_duration_minutes'] = df['trip_duration_seconds'] / 60
    
    # Display the DataFrame with trip duration columns
    print(f"DataFrame: {df_name}")
    print(df[['started_at', 'ended_at', 'trip_duration_seconds', 'trip_duration_minutes']])
    print("\n")
```

```python
df1 = df1.drop(columns=['ride_id', 'rideable_type', 'member_casual'])
df2 = df2.drop(columns=['ride_id', 'rideable_type', 'member_casual'])
df3 = df3.drop(columns=['ride_id', 'rideable_type', 'member_casual'])
df4 = df4.drop(columns=['ride_id', 'rideable_type', 'member_casual'])

```

```python
# Read data from 2014 and 2018 July files
file5 = '14_July_1.csv'
file6 = '18_July_1.csv'
file7 = '18_July_2.csv'


df5 = pd.read_csv(file5)
df6 = pd.read_csv(file6)
df7 = pd.read_csv(file7)
```

```python
# List of DataFrame names and the corresponding DataFrame objects
dfs2 = {'df5': df5, 'df6': df6, 'df7': df7}

for df_name, df in dfs2.items():
    # Convert 'started_at' and 'ended_at' to datetime if they are not already
    df['starttime'] = pd.to_datetime(df['starttime'])
    df['stoptime'] = pd.to_datetime(df['stoptime'])
    
    # Calculate the time difference in seconds
    df['trip_duration_seconds'] = (df['stoptime'] - df['starttime']).dt.total_seconds()
    
    # Optionally, create a new column for trip duration in minutes
    df['trip_duration_minutes'] = df['trip_duration_seconds'] / 60
    
    # Display the DataFrame with trip duration columns
    print(f"DataFrame: {df_name}")
    print(df[['starttime', 'stoptime', 'trip_duration_seconds', 'trip_duration_minutes']])
    print("\n")
```

```python
df5 = df5.drop(columns=['bikeid','usertype', 'birth year', 'gender', 'tripduration'])
df6 = df6.drop(columns=['bikeid','usertype', 'birth year', 'gender', 'tripduration'])
df7 = df7.drop(columns=['bikeid','usertype', 'birth year', 'gender', 'tripduration'])

```
